```python
from openai import OpenAI
client = OpenAI()
```

```python
def extraction_agent(ehr_text):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a medical data extraction agent. Extract structured information in JSON format."},
            {"role": "user", "content": ehr_text}
        ]
    )
    return response.choices[0].message.content
```

```python
#Agent d’Anonymisation
def anonymization_agent(structured_data):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a medical anonymization agent. Remove all personal identifiers."},
            {"role": "user", "content": structured_data}
        ]
    )
    return response.choices[0].message.content
```

```python
#Agent d’Analyse
def analysis_agent(anonymized_data):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a clinical analysis agent. Detect possible medical risks."},
            {"role": "user", "content": anonymized_data}
        ]
    )
    return response.choices[0].message.content
```

```python
#Agent de Résumé
def summary_agent(anonymized_data, analysis):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a medical summary agent. Produce a concise professional summary."},
            {"role": "user", "content": f"Medical Data:\n{anonymized_data}\n\nAnalysis:\n{analysis}"}
        ]
    )
    return response.choices[0].message.content
```

```python
#Agent de Coordination
def run_multi_agent_system(ehr_text):

    structured = extraction_agent(ehr_text)
    anonymized = anonymization_agent(structured)
    analysis = analysis_agent(anonymized)
    summary = summary_agent(anonymized, analysis)

    return {
        "Structured Data": structured,
        "Anonymized Data": anonymized,
        "Analysis": analysis,
        "Summary": summary
    }
```
